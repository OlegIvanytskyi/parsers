```python
from helper import *
driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
categories = {'politics' : 'politik', 'sport' : 'sport/fussball',
              'economy' : 'wirtschaft', 'career' : 'leben/karriere',
              'home' : 'leben/wohnen', 'health' : 'leben/gesundheit',
              'pets' : 'leben/tiere', 'gaming' : 'leben/games',
              'auto' : 'auto', 'travel' : 'reise',
              'cinema' : 'kino', 'food' : 'leben/genuss'}
website = 'https://www.merkur.de'
```

```python
for category, path in categories.items():
    print(f'Working with {category} category')
    
    lines = []
    
    for year in range(2002, 2021):
        print(f'Year: {year}')
        for month in range(1, 13):
            print(f'Month: {month}')
            page_num = 1
            month_str = f'0{month}' if month < 10 else str(month)

            while True:
                url = f'https://www.merkur.de/{path}/archiv-{year}-{month_str}/index-vc-376989-{page_num}.html#id-swiper-376989'

                driver.get(url)
                page = driver.page_source
                soup = BeautifulSoup(page, 'html.parser')

                if page_num == 1:
                    try:
                        limit = int(soup.find('div', class_='id-Swiper-navcount').text.split()[-1])
                    except AttributeError:
                        limit = 1
                    print(f'Number of pages: {limit}')

                div = soup.find('div', class_='id-LB id-LB--XL2 id-Teaser id-Teaser-el--listLeftStyle--LB')
                if div is None:
                    break

                divs = div.find_all('div', class_='id-LB-e--XL6_0c') + div.find_all('div', class_='id-LB-e--XL6_6')
                if divs == []:
                    break

                for div in divs:
                    teaser_headline = div.find('h3', class_='id-Teaser-el-content-headline')
                    teaser = div.find('div', class_='id-Teaser-el-content-text')

                    if None in (teaser_headline, teaser):
                        continue

                    # going to the page of the article to get keywords
                    href = div.find('a', class_='id-LinkOverlay-link')['href']
                    try:
                        article = driver.get(website + href)
                        article_soup = BeautifulSoup(driver.page_source, 'html.parser')                        
                        tags = article_soup.find_all('meta', attrs={'property':'article:tag'})
                        keywords = ' '.join([tag['content'] for tag in tags])
                    except TypeError:
                        keywords = ''

                    teaser_headline = teaser_headline.text.strip()
                    teaser = teaser.text.strip()

                    lines.append(teaser_headline)
                    lines.append(teaser)
                    lines.append(keywords)
                        
                print(f'Scraped {len(lines)} lines')

                page_num += 1
                if page_num > limit:
                    print()
                    break
                
    print(f'Finished working with {category} category. Scraped {len(lines)} lines\n')
    write_to_file(lines, category, 'merkur_de')
```
