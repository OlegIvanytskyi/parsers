```python
categories = {'pets' : 'haustiere', 'food' : 'essen-und-trinken',
              'digital' : 'software', 'gaming' : 'computerspiele',
              'nature' : 'nachhaltigkeit-natur', 'parenting' : 'eltern',
              'career' : 'karriere', 'politics' : 'rassismus',
              'cinema' : 'kinofilme', 'realestate' : 'immobilienmarkt'}
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())

for category, path in categories.items():
    print(f'Working with {category} category')

    articles = []
    page_num = 1

    while True:
        url = f'https://www.welt.de/themen/{path}/{page_num}/'
        driver.get(url)
        page = driver.page_source

        soup = BeautifulSoup(page, 'html.parser')
        
        if page_num == 1:
            try:
                limit = int(soup.find('ul', class_='c-pagination').find_all('li')[-2].text)
            except AttributeError:
                limit = 1
            print(f'Number of pages: {limit}')
        
        main = soup.find('main', class_='c-page-container c-page-teaser-compact-view')

        if main is None:
            break

        lis = main.find_all('li', class_='c-grid__item c-grid__item--is-one-third')

        for li in lis:
            teaser_headline = li.find('div', class_='c-dreifaltigkeit c-teaser-default__dreifaltigkeit')
            teaser = li.find('div', class_='c-teaser-default__intro o-teaser__intro o-text')

            if None in (teaser_headline, teaser):
                continue

            teaser_headline = teaser_headline.text.strip()
            teaser = teaser.text.strip()
            
            articles.append(teaser_headline)
            articles.append(teaser)

            if len(articles) % 50 == 0:
                print(len(articles))

        page_num += 1
        if page_num > limit:
            break
            
    print(f'Finished working with {category} category. Scraped {len(articles)} articles\n\n')
    write_to_file(articles, category, 'welt_de')
```
